```python
import numpy as np 
from tensorflow import keras
from tensorflow.keras import layers
```

```python
training_data, testing_data = keras.datasets.cifar10.load_data()
```

```python
X_train, y_train = training_data
X_test, y_test = testing_data
```

```python
X_train.shape
y_train.shape

y_train
```

```python
X_train, X_test = X_train/255.0, X_test/255.0
X_train
```

```python
X_train
```

```python
y_train = keras.utils.to_categorical(y_train, num_classes=10)
y_test = keras.utils.to_categorical(y_test, num_classes=10)
```

```python
y_train
```

```python
model = keras.Sequential(
    [
        layers.Conv2D(32, (3,3), activation='relu', input_shape=(32, 32, 3)),
        layers.MaxPooling2D((2,2)),
        layers.Conv2D(64, (3,3), activation='relu'),
        layers.MaxPooling2D((2,2)),
        layers.Conv2D(64, (3,3), activation='relu'),
        layers.Flatten(),
        layers.Dense(64, activation='relu'),
        layers.Dense(10, activation='softmax')
    ]
)

model.compile(
    optimizer='adam',
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

model.fit(X_train, y_train, epochs=30, validation_data=(X_test, y_test))

test_loss, test_acc = model.evaluate(X_test, y_test)
print(f'Teast Accuracy: {test_acc:2f}')
```

```python
fashion_mnist = keras.datasets.fashion_mnist
```

```python
(X_train_new, y_train_new), (X_test_new, y_test_new) = fashion_mnist.load_data()

X_validation, X_main_train = X_train_new[:5000] / 255.0, X_train_new[5000:] / 255.0
y_validation, y_main_train = y_train_new[:5000], y_train_new[5000:]


fashion_model = keras.models.Sequential([
    keras.layers.Flatten(input_shape=[28, 28]),
    keras.layers.Dense(300, activation="relu"),
    keras.layers.Dense(100, activation="relu"),
    keras.layers.Dense(10, activation='softmax')
])

fashion_model.summary()

print(X_test_new.shape)
```
